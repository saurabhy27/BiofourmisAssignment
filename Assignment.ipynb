```python
import pandas as pd
from datetime import datetime
import calendar
import time
import os
```

```python
df = pd.read_csv("input.csv")
```

```python
df['datetime'] = pd.to_datetime(df['timestamp'], unit='s')
```

```python
df1 = df[['datetime', 'heart_rate', 'respiration_rate', 'activity']].groupby([df['datetime'].dt.round('3600S')]).agg(['mean', 'min', 'max']).reset_index(drop=True)
```

```python
def convert_into_timestamp(timestamp):
    return calendar.timegm(timestamp.to_pydatetime().timetuple())
```

```python
df1['seg_start'] = df1.loc[:, 'Datetime']['min'].apply(convert_into_timestamp)
df1['seg_end'] = df1.loc[:, 'Datetime']['max'].apply(convert_into_timestamp)
df1.drop(('Datetime', 'mean'), axis=1, inplace = True)
df1.drop(('Datetime', 'min'), axis=1, inplace = True)
df1.drop(('Datetime', 'max'), axis=1, inplace = True)
```
